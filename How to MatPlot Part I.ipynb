```python
#import matplotlib

import matplotlib.pyplot as plt
import numpy as np
```

# Pyplot

Draw a line in a diagram from position (0,0) to position (6,250):



```python
xpoints = np.array([0, 10])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1, 11])
ypoints = np.array([1, 300])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#add 'o' to make a pyplot without a line

xpoints = np.array([1, 11])
ypoints = np.array([1, 300])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
#Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):

xpoints = np.array([1,3,10,14])
ypoints = np.array([50,100,50,250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points. 

ypoints = np.array([3,8,1,10,5,7])

plt.plot(ypoints)
plt.show()
```

# Markers

You can use the keyword argument `marker` to emphasize each point with a specified marker:

```python
ypoints = np.array([3,8,1,10,5,7])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
# change 'o'to '*' (asterisk) to make a marker into a star

ypoints = np.array([3,8,1,10,5,7])

plt.plot(ypoints, marker = '*')
plt.show()
```

## Format Strings `fmt`

You can also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this syntax:

|`marker `| `line`| `color`|

```python

xpoints = np.array([1, 4, 6, 10])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, 'o:k') # o specifies the marker, : the line type, and b the color
plt.show()
```

#### Line Reference

|Line Syntax	|Description|
| --- | --- |
|'-'|	Solid line|	
|':'	|Dotted line	|
|'--'|	Dashed line|	
|'-.'|	Dashed/dotted line|

#### Color Reference

|Color Syntax|	Description|
|---|---|
|'r'|	Red|	
|'g'|	Green|	
|'b'|	Blue|	
|'c'|	Cyan|	
|'m'|	Magenta|	
|'y'|	Yellow|	
|'k'|	Black|	
|'w'|	White|

```python
xpoints = np.array([1, 4, 6, 10])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, 'o-.m') 
plt.show()
```

## Marker Size

You can use the keyword argument `markersize` or the shorter version, `ms` to set the size of the markers:

```python
xpoints = np.array([1, 4, 6, 10])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, 'o-.m', ms = 20) 
plt.show()
```

## Marker Color

You can use the keyword argument `markeredgecolor` or the shorter `mec` to set the color of the edge of the markers:



```python
xpoints = np.array([1, 4, 6, 10])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, 'o-.r', ms = 20, mec = 'y') 
plt.show()
```

You can use the keyword argument `markerfacecolor` or the shorter `mfc` to set the color inside the edge of the markers:



```python
xpoints = np.array([1, 4, 6, 10])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, 'o-.r', ms = 20, mec = 'y', mfc = 'y') 
plt.show()
```

## Color Options

You can also use [Hexadecimal](https://www.w3schools.com/colors/colors_hexadecimal.asp) color values  
Or any of the [140 supported color](https://www.w3schools.com/colors/colors_names.asp) names.

```python
xpoints = np.array([1, 4, 6, 10])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, color = 'Darkslategrey') 
plt.show()
```

# Linestyle

You can use the keyword argument `linestyle`, or shorter `ls`, to change the style of the plotted line:



```python
#dotted line
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted', marker= '*', color = 'hotpink')
plt.show()
```

```python
#dashed line
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dashed', marker= '*', color = 'hotpink')
plt.show()
```

#### Shorter Syntax
The line style can be written in a shorter syntax:

|Style| Shorter|
| --- | --- |
|linestyle| ls|
|dotted| : |
| dashed | -- |

## Line Color 

You can use the keyword argument `color` or the shorter `c` to set the color of the line:


```python
ypoints = [0, 3, 5, 10]

plt.plot (ypoints, marker = 'x', c = 'hotpink')
plt.show()
```

[COLOR REFERENCE](#Color-Options)

## Line Width

You can use the keyword argument `linewidth` or the shorter` lw` to change the width of the line.



```python
ypoints = [0, 3, 5, 10]

plt.plot (ypoints, lw = 20, c = 'lightgoldenrodyellow')
plt.show()
```

## Multiple Lines

You can plot as many lines as you like by simply adding more `plt.plot()` functions:



```python
y1 = np.array([3, 8, 1, 10])
x1 = np.array([0, 50, 100, 150])
y2 = np.array([6, 2, 7, 11])
x2 = np.array([0, 50, 100, 150])

plt.plot(x1,y1, c = 'r', marker = 'o')
plt.plot(x2, y2, c = 'forestgreen', marker = 'o')

plt.show()
```
